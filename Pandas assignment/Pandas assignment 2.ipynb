# Assignments on HR Data
    convert average_montly_hours into days and hours
    give hike to people still with company and did in top 25% of performance

```python
import pandas as pd
hr_data=pd.read_csv('https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/HR_comma_sep.csv.txt')
hr_data.head()
```

```python
def f(d):
    day=d.average_montly_hours//24
    hour=d.average_montly_hours%24
    return "{} days {} hours".format(day,hour)
hr_data["Average_monthly_days"]=hr_data.apply(f,axis=1)
hr_data.head()
```

##give hike to people still with company and did in top 25% of performance

```python
hr_data.describe()
```

```python
def f(d):
    if d.last_evaluation>0.87 and d.left==0:
        if d.salary=='low':
            return 'medium'
        elif d.salary=='medium':
            return 'high'
        else:
            return 'very high'
    else:
        return d.salary
    
hr_data["new_salary"]=hr_data.apply(f,axis=1)
hr_data.head()
```

```python
hr_data[hr_data.salary!= hr_data.new_salary].head()
```

# Assignments on Titanic Data
create a new column with values KID, ADULT and & OLD based on age
find the richest person

```python
titanic=pd.read_csv('https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/titanic-train.csv.txt')
titanic.head()
```

```python
def f(a):
    if a.Age<16:
        return "kid"
    elif a.Age>15 and a.Age<60:
        return "adult"
    else:
        return "old"
    
titanic["category"]=titanic.apply(f, axis=1)
titanic.head()
```

```python
titanic[titanic.category=='kid'].head(20)
```

# Assignments on Titanic Data

find the richest person

```python
max=0
for fare in titanic.Fare:
    if max<fare:
        max=fare
print(titanic[titanic.Fare==max]["Name"])
```
